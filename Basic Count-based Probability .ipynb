# Question 3

Simple random walk

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed to ensure consistent results each time the code is run
np.random.seed(42)

# Define the number of steps in the random walk
num_steps = 1000

# Generate an array of random steps, either +1 or -1 with equal probability
steps = 2*np.random.randint(2, size=num_steps)-1

# Calculate the cumulative sum of the steps to get the random walk
walk = np.cumsum(steps)

# Plot the random walk
plt.plot(walk)

# Add title and labels to the plot
plt.title("Simple Random Walk")
plt.xlabel("Step")
plt.ylabel("Position")

# Show the plot
plt.show()
```
