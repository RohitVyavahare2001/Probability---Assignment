# Question 1


Write a python program to plot the normal distribution function with μ = 0.5 and Σ =1



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

u = 0
sigma = 1
x = np.linspace(-5, 5, 1000)
y = stats.norm.pdf(x, u, sigma)

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Normal Distribution with μ = 0 and Σ = 1")
plt.show()
```

# Question 7


 Calculating Probability of Specific Data Occurrence, if we were asked to pick one person randomly from this distribution, then what is the probability that the height of the person will be smaller than 4.5 ft. ?


```python
import numpy as np

# create a random dataset of heights
heights = np.random.normal(5, 1, 1000)

# calculate the probability of picking a person with height smaller than 4.5 ft
prob = sum(heights < 4.5) / len(heights)

print(f"The probability of picking a person with height smaller than 4.5 ft is {prob:.2f}")
```

# Question 8
pick one person randomly from this distribution, then what is the probability that the height
of the person will be between 6.5 and 4.5 ft. ?
Area Under The Curve Between 4.5 And 6.5 Ft


```python
import scipy.stats as stats

mu = 5
sigma = 1.5

# calculate the CDF at 4.5 ft
cdf_at_4_5 = stats.norm.cdf(4.5, loc=mu, scale=sigma)

# calculate the CDF at 6.5 ft
cdf_at_6_5 = stats.norm.cdf(6.5, loc=mu, scale=sigma)

# calculate the probability of height being between 6.5 and 4.5 ft
prob = cdf_at_6_5 - cdf_at_4_5
print("Probability of height between 6.5 and 4.5 ft: ", prob)
```

# Question 9
Probability Density
what if we were asked about the probability that the height of a person chosen randomly
will be above 6.5ft?

```python
import scipy.stats as stats

# assuming the mean and standard deviation of the distribution are known
mu = 5
sigma = 1.5

# calculate the CDF at 6.5 ft
cdf_below_6_5 = stats.norm.cdf(6.5, loc=mu, scale=sigma)

# subtract the CDF from 1 to get the probability of height above 6.5ft
prob_above_6_5 = 1 - cdf_below_6_5

print("Probability of height above 6.5ft:", prob_above_6_5)
```
